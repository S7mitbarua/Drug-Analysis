```python
import numpy as np
import pandas as pd
import random

# Set a random seed for reproducibility
np.random.seed(42)

# Simulate real-world prescription data
num_patients = 500
patient_ids = np.arange(1, num_patients + 1)

# Define drug classes with associated medications and health conditions
drug_classes = {
    'PainRelievers': ['Aspirin', 'Ibuprofen', 'Acetaminophen'],
    'Antibiotics': ['Amoxicillin', 'Ciprofloxacin', 'Azithromycin'],
    'Antidepressants': ['Sertraline', 'Fluoxetine', 'Escitalopram'],
}

health_conditions = ['Pain', 'Infection', 'Depression']

# Simulate prescription records for each patient
prescriptions = []

for patient_id in patient_ids:
    for _ in range(random.randint(1, 5)):
        drug_class = random.choice(list(drug_classes.keys()))
        drug = random.choice(drug_classes[drug_class])
        condition = random.choice(health_conditions)
        adherence = np.random.uniform(0.5, 1.0)
        side_effect = random.choice(['Headache', 'Nausea', 'None'])

        prescriptions.append({
            'PatientID': patient_id,
            'DrugClass': drug_class,
            'Drug': drug,
            'HealthCondition': condition,
            'Adherence': adherence,
            'SideEffect': side_effect,
        })

# Create a DataFrame from the simulated data
prescription_data = pd.DataFrame(prescriptions)

# Display the first few rows of the simulated real-world substance data
prescription_data.head()
```
